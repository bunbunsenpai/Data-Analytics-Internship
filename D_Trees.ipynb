```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('movies.csv')
df
```

```python
sns.scatterplot(data=df, x='Celebrities', y='Budget')
```

```python
sns.jointplot(data=df, x='Celebrities', y='Budget', hue='Class')
```

```python
x=df[['Budget','Celebrities']]

y=df['Class']

x
```

```python
##Build The Model
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier(random_state=0)
```

```python
dt.fit(x,y)
```

```python
#Predict On New Values
```

```python
new1=[[5000000,3]]
```

```python
dt.predict(new1)
```

```python
new2=[[5000000,8]]
```
