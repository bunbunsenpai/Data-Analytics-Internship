```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python


x=np.array([12,23,34,45,49,56,61,68,71,78])

y=np.array([23,31,38,41,45,51,61,69,70,75])

z=np.array([16,19,25,35,45,55,60,65,69,79])
```

Line Plot

```python
plt.title('Death Rate')
plt.xlabel('Time')
plt.ylabel('People')
plt.plot(x, y, marker= '>', color ='gold', mfc='r' , ls='--' ,ms='6' );

plt.grid()
```

```python
plt.title('Death Rate')
plt.xlabel('Time')
plt.ylabel('People')
plt.plot(x, z, marker= 'o', color ='blue', mfc='y' , ls='--' ,ms='4' );

plt.grid()
```

```python
plt.title('Death Rate')
plt.xlabel('Time')
plt.ylabel('People')
plt.plot(x, z, marker= 'o', color ='silver', mfc='y' , ls='--' ,ms='4' );
plt.plot(x, y, marker= '>', color ='gold', mfc='r' , ls='--' ,ms='6' );

plt.grid()
```

Subplot


```python
plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.title('Death Rate')
plt.xlabel('Time')
plt.ylabel('People')
plt.plot(x, z, marker='o', color='blue', mfc='y', ls='--', ms='4')

plt.grid()

plt.subplot(2, 2, 2)
plt.title('Death Rate')
plt.xlabel('Time')
plt.ylabel('People')
plt.plot(x, y, marker='>', color='gold', mfc='r', ls='--', ms='6')

plt.grid()
```

Bar Graph


```python
students=[50,60,41,45,21]
courses=['java','css','cc','ai','wt']
```

```python
plt.title('Enrollment')
plt.xlabel('Courses')
plt.ylabel('No. Of Students')
plt.bar(courses, students, color='Black', edgecolor='gold', width= 0.5, linewidth=1 , alpha=1);
```

Scatter Plot

```python
plt.scatter(x,y , marker='o',color='purple');
plt.scatter(z,y , marker='o',color='blue');
plt.scatter(x,z, marker='o',color='gold');

plt.grid()
```

Pie Chart

```python
plt.title('Pie Chart')
plt.pie(students, labels=courses ,autopct='%.2f%%', shadow= True, radius=1
       , labeldistance=1.2,colors=['blue','red','green', 'pink','gold'],startangle=30 ,
explode=[0,0.2,0,0,0] );
```

```python
plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.title('Pie Chart')
plt.pie(students, labels=courses ,autopct='%.2f%%', shadow= True, radius=1
       , labeldistance=1.2,colors=['blue','red','green', 'pink','gold'],startangle=30 ,
explode=[0,0.2,0,0,0] );

plt.subplot(2, 2, 2)
plt.scatter(x,y , marker='o',color='purple');
plt.scatter(z,y , marker='o',color='blue');
plt.scatter(x,z, marker='o',color='gold');

plt.grid()

plt.subplot(2, 2, 3)
plt.title('Enrollment')
plt.xlabel('Courses')
plt.ylabel('No. Of Students')
plt.bar(courses, students, color='Black', edgecolor='gold', width= 0.5, linewidth=1 , alpha=1);


plt.subplot(2, 2, 4)
plt.title('Death Rate')
plt.xlabel('Time')
plt.ylabel('People')
plt.plot(x, z, marker= 'o', color ='silver', mfc='y' , ls='--' ,ms='4' );
plt.plot(x, y, marker= '>', color ='gold', mfc='r' , ls='--' ,ms='6' );

plt.grid()

plt.savefig('My Plot')
```

Box Plot

```python
marks=[67,76,50,65,72,76,81,12,85,120]
```

```python
plt.boxplot(marks);
```

```python
q3=np.percentile(marks,85)
q1=np.percentile(marks,50)
```

```python
q3,q1
```

```python
iqr=q3-q1
iqr #interquartile range
```

```python
#lowerbound
lb=q1-1.5*iqr

ub=q3+1.5*iqr
```

```python
lb,ub
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)
```

```python
sns.jointplot(x='tip', y='total_bill', data=tips, hue='time')
plt.show()
```

Histogram

```python
plt.xlabel('Tip Amount')
plt.ylabel('Frequency')
sns.histplot(tips['tip'],kde=True, color='Gold')
plt.show()
```

```python
plt.xlabel('day')
plt.ylabel('tip')
sns.boxenplot(x='day',y='tip',data=tips, color='red')
plt.show()
```

```python
sns.pairplot(tips,hue='smoker')
plt.show()
```
