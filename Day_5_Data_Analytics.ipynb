```python
q1=np.percentile(df['salary'],25)
q3=np.percentile(df['salary'],75)
```

```python
q1,q3
```

```python
iqr=q3-q1
```

```python
upperfence = q3+iqr*1.5
lowerfence = q1-iqr*1.5
```

```python
upperfence,lowerfence
```

```python
ndf=df[(df['salary'] < upperfence) & (df['salary'] > lowerfence)]
```
